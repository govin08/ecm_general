```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터 불러오기
path_root = os.path.dirname(os.path.abspath('.'))
# data = pd.read_csv(os.path.join(path_root, 'data', 'data0904.csv'))
data = pd.read_csv(os.path.join(path_root, 'data', 'mdf_data', 'seasonal', 'data_jan4_5s_7d.csv'), encoding='cp949')
data.rename(columns={'연료대비 질소가스 주입 비율':'희석질소 오프셋',
             'DGAN Compressor 질소가스 유량':'희석질소 유량',
             '후단 질소산화물 농도':'NOx 농도'}, inplace=True)
data
```

```python
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

offset_name = '희석질소 오프셋'
action_name = '희석질소 유량'
target_name = 'NOx 농도'

plt.rcParams['font.size'] = 14  # 기본 폰트 크기
plt.rcParams['axes.labelsize'] = 16  # 축 레이블 크기
plt.rcParams['legend.fontsize'] = 14  # 범례 폰트 크기
plt.rcParams['xtick.labelsize'] = 12  # x축 눈금 레이블 크기
plt.rcParams['ytick.labelsize'] = 12  # y축 눈금 레이블 크기

def plot(name):
    plt.figure(figsize=(20,5))
    plt.plot(data[name], label=name)
    plt.axvline(x=0.8*len(data), color='red', linestyle='--', alpha=0.7,)
    plt.legend()
    plt.show()

plot(offset_name)
plot(action_name)
plot(target_name)
```
